```python
import pandas as pd
import numpy as np
```

```python
path="/content/boiling-32_temp_heat_flux.txt"
df=pd.read_csv(path, sep='\t')
df=pd.read_csv(path, sep='\s+')
df
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
print(df.columns)
x=df['Temperature(C)'].values
y=df['Heat_flux(W/cm2)'].values
plt.plot(x,y)
plt.xlabel('Temperature(C)',color='purple',fontsize='20')
plt.ylabel('Heat_flux(W/cm2)',color='purple',fontsize='20')
plt.show()
```

Finding value of m and c of y=mx+c  

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Temperature(C)']],df['Heat_flux(W/cm2)'])
print(reg.coef_)
print(reg.intercept_)
```

```python
reg.predict([[150]])
```

Now we will find the predicted value only

```python
x=df[['Temperature(C)']]
d=reg.predict(x)
df['Prediction']=d
```

```python
df
```

```python
x=df['Temperature(C)'].values
y=df['Heat_flux(W/cm2)'].values
z=df['Prediction'].values
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('Temperature(C)',color='purple',fontsize='20')
plt.ylabel(' Heat_flux(W/cm2)',color='purple',fontsize='20')
plt.show()
```

```python
r_squared = reg.score(df[['Temperature(C)']], df['Heat_flux(W/cm2)'])
r_squared
```
