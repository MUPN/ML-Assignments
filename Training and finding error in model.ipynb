We can eaily fit the whole dataset but this won't make our model pefect .So we have to differenciate the data in 2 part.Train and Validation.

I haven't learn to fit string so i will work with the numerical ones.

```python
feat=["Bedroom2","Bathroom","Car","Lattitude","Distance"]
x=ml_model[feat]
x
```

```python
y=ml_model["Price"]
y
```

Taking metrics to find error and model selction to train the model

```python
from sklearn.model_selection import train_test_split

train_x,v_x,train_y,v_y=train_test_split(x,y,random_state=0)

ml_model.fit(train_x,train_y)
```

```python
from sklearn.tree import DecisionTreeRegressor
ml_model=DecisionTreeRegressor(random_state=1)
ml_model.fit(train_x,train_y)
```

```python
from sklearn.metrics import mean_absolute_error
pred_y=ml_model.predict(v_x)
pred_y
```

```python
mean_absolute_error(v_y,pred_y)
```
